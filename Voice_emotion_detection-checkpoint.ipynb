## Importing Modules

```python
import glob
import os
import librosa
import librosa.display
import numpy as np
import pandas as pd
import _pickle as pickle
from sklearn import svm
from sklearn import model_selection
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import NearestNeighbors
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasClassifier
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Model
from tensorflow.keras.utils import plot_model
from tensorflow.keras.optimizers import SGD, Adadelta, Adagrad
from tensorflow.keras.layers import Dropout
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import plot_model
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from IPython.display import SVG
import seaborn as sn
from matplotlib import pyplot as plt
```

## Extracting Features from the Audio

```python
def FeatureExtraction(file_name):
     """Extract audio features from a given file
        
        Parameters:
        file_name (str):- Path to the audio file"""
    
    # Load the audio file with the original sample rate
    X, sample_rate = librosa.load(file_name, sr=None)

    # Compute the Short-Time Fourier Transform (STFT)
    stft = np.abs(librosa.stft(X))

    # Extract Mean Mel-Frequency Cepstral Coefficients (MFCCs)
    mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T, axis=0)

    # Extract Chroma feature
    chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T, axis=0)

    # Extract Mel-scaled spectrogram
    mel = np.mean(librosa.feature.melspectrogram(y=X, sr=sample_rate).T, axis=0)

    # Extract Spectral Contrast
    contrast = np.mean(librosa.feature.spectral_contrast(S=stft, sr=sample_rate).T, axis=0)
    
    # Extract Tonnetz (tonal centroid features)
    tonnetz = np.mean(librosa.feature.tonnetz(y=librosa.effects.harmonic(X), sr=sample_rate).T, axis=0)

'''Returns:- tuple:- Contains extracted features (MFCCs, Chroma, Mel, Contrast, Tonnetz)'''
    return mfccs, chroma, mel, contrast, tonnetz
           
```

```python
def AudioFilesParsing(path):
     """Parse audio files to extract features and labels.
    
    Parameters:- path (str): Glob pattern for the file paths."""

    features = []
    labels = []
    for ft in glob.glob(path):
        result = FeatureExtraction(ft)
        if result is not None:
            mfccs, chroma, mel, contrast, tonnetz = result
            ExtractedFeatures = np.hstack([mfccs, chroma, mel, contrast, tonnetz])
            features.append(ExtractedFeatures)
            
            label = os.path.basename(ft).split("_")[2] 
            labels.append(label)
                
     '''Returns:
    tuple: Contains feature array and label array.'''       
    return np.array(features), np.array(labels)

# Define the path pattern for training files
path = r".\train*\*\*.wav"
X, Y = AudioFilesParsing(path)
# Ensuring the number of samples in features and labels match
assert X.shape[0] == len(Y)
```

```python
# Parse the training audio files to extracted features and labels
TrainFeatures, TrainLabels = AudioFilesParsing('./train*/*/*.wav')

# Converting lists of features and labels to pandas Series and then to numpy arrays
TrainFeatures = np.array(TrainFeatures, dtype=pd.Series)
TrainLabels = np.array(TrainLabels, dtype=pd.Series)

X = TrainFeatures.astype(int)   # Features as integers
Y = TrainLabels.astype(str)     # Labels as strings

# Checking the shapes of the feature and label arrays
X.shape, Y.shape
```

```python
# Random seed for reproducibility
seed = 9
np.random.seed(seed)
# Encoding class values as integers
encoder = LabelEncoder()
encoder.fit(Y)
YEncoded = encoder.transform(Y)

Y_Demo = to_categorical(YEncoded)
```

## Defining the Model

```python
def ModelBaseline():
    DeepModel = Sequential()
    DeepModel.add(Dense(100, input_dim=193, activation="relu", kernel_initializer="uniform"))
    DeepModel.add(Dropout(0.5))
    DeepModel.add(Dense(50, activation="relu", kernel_initializer="uniform"))
    DeepModel.add(Dropout(0.5))
    DeepModel.add(Dense(20, activation="relu", kernel_initializer="uniform"))
    DeepModel.add(Dropout(0.5))
    DeepModel.add(Dense(7, activation="softmax", kernel_initializer="uniform"))

    DeepModel.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return DeepModel
```

## Training the Model

```python
epochs = 150
batch_size = 25
verbose = 1

model = ModelBaseline()
history = model.fit(X, Y_Demo, 
                    validation_split=0.1,
                    batch_size=batch_size, 
                    epochs=epochs, 
                    verbose=verbose)

scores = model.evaluate(X, Y_Demo, verbose=0)
print(f"Validation Accuracy: {scores[1]*100:.2f}%")


print(history.history)

FileName = 'model_wieghts.wieght.h5'

model.save(FileName)

print('Model Saved...')
```

## Testing the model

```python
def ExtractFeature(FileName):
    X, SampleRate = librosa.load(FileName, sr=None)
    stft = np.abs(librosa.stft(X))
    mfccs = np.mean(librosa.feature.mfcc(y=X, sr=SampleRate, n_mfcc=40).T, axis=0)
    chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=SampleRate).T, axis=0)
    mel = np.mean(librosa.feature.melspectrogram(y=X, sr=SampleRate).T, axis=0)
    contrast = np.mean(librosa.feature.spectral_contrast(S=stft, sr=SampleRate).T, axis=0)
    tonnetz = np.mean(librosa.feature.tonnetz(y=librosa.effects.harmonic(X), sr=SampleRate).T, axis=0)
    return mfccs, chroma, mel, contrast, tonnetz
    
        
```

```python
def ParsingTestAudioFiles(path):
    features = []
    labels = []
    for fn in glob.glob(path):
        result = ExtractFeature(fn)
        if result is not None:
            mfccs, chroma, mel, contrast, tonnetz = result
            TestFeature = np.hstack([mfccs, chroma, mel, contrast, tonnetz])
            features.append(TestFeature)
            
            label = os.path.basename(fn).split("_")[2] 
            labels.append(label)
    return np.array(features), np.array(labels)        

path = r".\test*\*\*.wav"
TestFeatures, TestLabels = ParsingTestAudioFiles(path)                
            
assert TestFeatures.shape[0] == len(TestLabels)   

encoder = LabelEncoder()
encoder.fit(TestLabels.astype(str))
Encoded_Y = encoder.transform(TestLabels.astype(str))
TestLabels = to_categorical(Encoded_Y)

TestLabels = TestLabels.reshape(TestLabels.shape[0], 7)

model = load_model('model_wieghts.wieght.h5')

prediction = model.predict(TestFeatures.astype(float))

PredictedIndices = np.argmax(prediction, axis=1)

TrueIndices = np.argmax(TestLabels, axis=1)

accuracy = accuracy_score(TrueIndices, PredictedIndices)


# Mapping predicted class indices to class labels
LabelsMap = {0: 'Angry', 1: 'Disgusted', 2: 'Feared', 3: 'Happy', 4: 'Neutral', 5: 'Pleasant Surprise', 6: 'Sad'}
TestPredicted = [LabelsMap[Ids] for Ids in PredictedIndices]

# Ensuring true labels are also mapped 
TestTrue = [LabelsMap[Ids] for Ids in TrueIndices]

# Calculating accuracy
accuracy = accuracy_score(TestTrue, TestPredicted)
print("Accuracy Score of Model:", accuracy)
print('Number of correct predictions made:', accuracy_score(TestTrue, 
                                                      TestPredicted, normalize=False),
      'out of', len(TestTrue))
```

## Visualising the Accuracy of Model
